```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt,seaborn as sns
```

```python
food_type=["Cooked rice dish","Non-veg curries","Dairy-based curries","Dal/lentils","Gravy-based veg curries","Fresh breads","Dry vegetable dishes","Fried items","Sweets"]
```

```python
temperature=['refrigerated temperature','room temperature(more than 25 C)','cool temperature(approx. 18 C)']
```

```python
row=[]
```

```python
for i in range(0,10000):
  food=np.random.choice(food_type)
  temp=np.random.choice(temperature)
  hours_already_spent=round(np.random.uniform(0,12),2)
  if temp=="refrigerated temperature":
    if food=="Cooked rice dish" or food=="Fresh breads":
      spoilage_time=48
    elif food=="Non-veg curries" or food=="Dairy-based curries" or food=="Dal/lentils" or food=="Gravy-based veg curries" or food=="Sweets" or food=="Fried items":
      spoilage_time=72
    elif food=="Dry vegetable dishes":
      spoilage_time=96
  elif temp=="room temperature(more than 25 C)":
    if food=="Cooked rice dish" or food=="Dairy-based curries":
      spoilage_time=3
    elif food=="Non-veg curries":
      spoilage_time=2
    elif food=="Dal/lentils" or food=="Gravy-based veg curries":
      spoilage_time=4
    elif food=="Dry vegetable dishes":
      spoilage_time=6
    elif food=="Fresh breads" or food=="Fried items":
      spoilage_time=8
    elif food=="Sweets":
      spoilage_time=5
  elif temp=="cool temperature(approx. 18 C)":
    if food=="Cooked rice dish" or food=="Dairy-based curries" or food=="Non-veg curries":
      spoilage_time=4
    elif food=="Dal/lentils" or food=="Gravy-based veg curries":
      spoilage_time=8
    elif food=="Dry vegetable dishes" or food=="Fresh breads" or food=="Sweets":
      spoilage_time=12
    elif food=="Fried items":
      spoilage_time=24
  rem_time=spoilage_time-hours_already_spent
  if rem_time<2:
    risk="very high"
  elif rem_time>=2 and rem_time<4:
    risk="high"
  elif rem_time>=4 and rem_time<8:
    risk="medium"
  elif rem_time>=8:
    risk="low"
  row.append([food,temp,hours_already_spent,spoilage_time,rem_time,risk])
```

```python
dt=pd.DataFrame(row,columns=["food_type","temperature","hours already spent","spoilage_time","remaining time","risk"])
```
